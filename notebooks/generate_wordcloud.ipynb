# Reddit

```python
# read in a csv file
os.chdir("C:/Users/Gary/Desktop/Year 1 Sem 2/CS5246 Text Mining/Group Project/depression-detector/data/test/")
file = "reddit_golden_test.csv"
df_orig = pd.read_csv(file)  
```

```python
#show first 5 rows
df_orig.head()
```

```python
df_orig.info()
```

```python
df_orig['cleanedreddit'].head()
```

```python
#generate wordcloud 
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

for i in range(0,3):
    df_selected = df_orig[df_orig['label'] == i]
    stopwords = set(STOPWORDS)
    stopwords.update(["https", "co", "reddit", "com", "cleanedreddit"])

    wordcloud = WordCloud(
                                background_color='white',
                                stopwords=stopwords,
                                max_words=200,
                                max_font_size=40,
                                random_state=42
                                ).generate(str(df_selected['cleanedreddit']))

    print(wordcloud)

    fig = plt.figure(1)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    fig.savefig("reddit_wordcloud" + str(i) + ".png", dpi=900)
```
